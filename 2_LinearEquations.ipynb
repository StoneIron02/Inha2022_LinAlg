```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def draw_linear_graph(x, m, n):
  y = [(m * num + n) for num in x]
  return y;
```

```python
x = np.arange(-1, 4, 0.1)
plt.plot(x, draw_linear_graph(x, -1, 2), x, draw_linear_graph(x, 1, -1))
plt.show()
```

```python
plt.plot(x, draw_linear_graph(x, -1, 1), x, draw_linear_graph(x, 1, -1), x, draw_linear_graph(x, 1/3, 1))
plt.show()
```

```python
# 5x+2y=14
# 2x+y=4
a = [[5,2],[2,1]]
b = [14,4]
result = np.linalg.solve(a, b)
print(result)
```

```python
# 5x+2y+z=14
# 2x-3y-4z=4
# x+5y-2z=21
a = [[5,2,1],[2,-3,-4],[1,5,-2]]
b = [14,4,21]
result = np.linalg.solve(a, b)
print(result)
```
